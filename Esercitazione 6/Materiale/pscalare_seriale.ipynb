```python
%%writefile seriale.c
#include <stdio.h>
#include <stdlib.h>

int main() {
    struct timeval start, end;
    long seconds, useconds;
    double elapsed_time;

    int N = 1600000;


    printf("Dimensione dei vettori: %d\n", N);

    // Dichiarazione dei vettori
    float *vettore1 = (float *)malloc(N * sizeof(float));
    float *vettore2 = (float *)malloc(N * sizeof(float));

    // Inizializzazione del generatore di numeri casuali
    srand(time(NULL));

    // Generazione casuale degli elementi del primo vettore
    if(N < 100)
        printf("Elementi del primo vettore:\n");
    for(int i = 0; i < N; i++) {
        vettore1[i] = rand() % 100; // Numeri casuali tra 0 e 99
        if(N < 100)
            printf("%f ", vettore1[i]);
    }
    if(N < 100)
        printf("\n");

    // Generazione casuale degli elementi del secondo vettore
    if(N < 100)
        printf("Elementi del secondo vettore:\n");
    for(int i = 0; i < N; i++) {
        vettore2[i] = rand() % 100; // Numeri casuali tra 0 e 99
        if(N < 100)
            printf("%f ", vettore2[i]);
    }
    if(N < 100)
        printf("\n");

    gettimeofday(&start, NULL);  // Inizio del cronometro

    // Calcolo del prodotto scalare
    float prodotto_scalare = 0;
    for(int i = 0; i < N; i++) {
        prodotto_scalare += vettore1[i] * vettore2[i];
    }

    gettimeofday(&end, NULL);  // Fine del cronometro

    // Calcolo del tempo trascorso in millisecondi
    seconds = end.tv_sec - start.tv_sec;
    useconds = end.tv_usec - start.tv_usec;
    elapsed_time = seconds * 1000 + useconds / 1000.0;

    // Stampa del risultato
    printf("Il prodotto scalare dei due vettori è: %f\n", prodotto_scalare);
    printf("Tempo di esecuzione: %f ms\n", elapsed_time);

    return 0;
}
```
